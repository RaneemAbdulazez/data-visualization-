```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

1-choose data set , here I chose built in titanic data set to make life easier 

```python
sns.get_dataset_names()
titanic_df = sns.load_dataset("titanic")
titanic_df.head()
```

```python
still_alive=titanic_df.groupby('alive',as_index=False)['age'].mean()
still_alive
```

2-Plotting the survive data to compare how many people survived and how many were victms

```python
sns.countplot(data=titanic_df, x="survived")
```
